This example demonstrates the principle of resolution. We will try changing the resolution to see what features we can still identify, or resolve, in the image.

The image that we will load shows Cahokia Mounds (read more here: en.wikipedia.org/wiki/Cahokia). The mound on the right is roughly 30 m wide by 30 m long. 

We will explore two definitions of resolution:

1. Pixel size (most common definition)

2. Size of smallest feature that can be identified. This value could be similar to the pixel size, if the imagery you are using is very sensitive to the feature you are trying to identify, or much higher (2x or 3x higher, sometimes more) if the imagery you are viewing is not sensitive to the feature you are trying to identify.

In this exercise, you will first view a very high-resolution (0.15 m pixel size) image of Cahokia Mounds. Then you will experiment with the resolution (variable newres) to determine the coarsest pixel size needed to be able to identify the Cahokia mound on the right.

Load packages that we will be using

```
import matplotlib.image as img
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
import skimage.transform as st 
```

Load original image

```
field=img.imread('remote_sensing_course/rasters/cahokia_mounds.JPG',format='JPG')
```

Make function to change resolution of the raster

```
def change_resolution(raster,new_resolution,original_resolution=0.15):
  factor=original_resolution/new_resolution
  size_y=int(raster.shape[0]*factor)
  size_x=int(raster.shape[1]*factor)
  newrast=st.resize(raster,(size_y,size_x))
  return(newrast)
```

Set default plot size

```
plt.rcParams['figure.figsize'] = [16, 10]
plt.rcParams['figure.dpi'] = 100
```

Now for the fun part. You will experiment by changing the value of 'newres' (this sets the pixel size of the new image, in m) until you feel that you have chosen the maximum pixel size that enables you to still identify the Cahokia mound on the right. Record the pixel size, and the ratio of the length of the Cahokia mound on the right over the pixel size. 

```
newres=1               # this is the pixel size of the new image, in meters
new_image=change_resolution(field,newres)

plt.figure();
plt.subplot(2,2,1)
plt.imshow(field);plt.title('original image: 0.15 m pixel size')
plt.subplot(2,2,2);
plt.imshow(new_image);plt.title('new image: '+str(newres)+' m pixel size')
```
